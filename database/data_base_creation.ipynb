```python
from datetime import datetime, timedelta
import pandas as pd
def convert_days_to_date(days):
    """
    Convert a number of days into a date format.

    Parameters:
    days (int): The number of days to convert.

    Returns:
    str: The date in 'YYYY-MM-DD' format.
    """
    # Calculate the new date
    new_date = datetime.now() - timedelta(days=days)
    # Return the date formatted as 'YYYY-MM-DD'
    return new_date.strftime('%Y-%m-%d')

# Example usage:
days_input = 1  # Example: converting 10 days into a date
result_date = convert_days_to_date(days_input)
print(f"The date {days_input} days from today is: {result_date}")
```

```python
def fetch_recent_reviews(recent_days):
    try:
        conn = mysql.connector.connect(
            host='localhost',         # e.g., 'localhost'
            database='restaurant_review',
            user='root',     # e.g., 'root'
            password='MudassirL6'   # Your MySQL password
        )
        if conn.is_connected():
            # Calculate the date 10 days ago
            ten_days_ago = datetime.now() - timedelta(days=recent_days)
            print(ten_days_ago)
            query = "SELECT * FROM restaurant_complaints WHERE timestamp <= %s"
            cursor = conn.cursor()
            cursor.execute(query, (ten_days_ago,))
            result = cursor.fetchall()
            columns = [i[0] for i in cursor.description]
            reviews_df = pd.DataFrame(result, columns=columns)
            cursor.close()
            return reviews_df
        
        
    except Error as e:
        print(f"Error: {e}")
        return pd.DataFrame()  # Return an empty DataFrame in case of error
    finally:
        if conn.is_connected():
            conn.close()
```
